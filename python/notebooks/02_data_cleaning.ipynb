# Geolocation Cleaning

### Problem
The geolocation dataset contained duplicate records.

### Solution
Removed duplicate rows using pandas `drop_duplicates()`.

### Result
Created a cleaned dataset with only unique records for downstream analysis.

```python
df5.isnull().sum()
```

```python
df5[df5.isnull().any(axis=1)]
```

```python
df5[df5.isnull().any(axis=1)].shape
```

```python
df5_clean = df5.copy()

df5_clean["product_category_name"] = df5_clean["product_category_name"].fillna("Unknown")

numeric_cols = [
    "product_name_lenght",
    "product_description_lenght",
    "product_photos_qty",
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]

for col in numeric_cols:
    df5_clean[col] = df5_clean[col].fillna(df5_clean[col].median())
```

```python
df5_clean.isnull().sum()
```

```python
df5_clean.to_csv("../data/processed/products_clean.csv", index=False)
```

# Products Dataset Cleaning

## Issues Identified
- Missing values in product category and product attributes.

## Cleaning Performed
- Filled missing product categories with **"Unknown"**.
- Filled missing numeric attributes using the **median**, which is robust to outliers.

## Output
- Saved cleaned dataset as `products_clean.csv`.
